```
import pandas as pd
```

```
train=pd.read_csv('train.csv')
test=pd.read_csv('test.csv')
```

```
train.columns
```

```
data=train.drop(['price'],axis=1)
target=train['price']

print(data.shape,target.shape)
```

```
sel =['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot']
sel_data = data[sel]
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(sel_data,target, random_state=42)
```

```
from sklearn.ensemble import RandomForestRegressor
# randomforestregressor 객체 만들기
# 트리의 객수를 5개로 한다.
model = RandomForestRegressor(n_estimators=5, random_state=2)
# 모델 학습시키기
model = model.fit(X_train, y_train)
```

```
model.feature_importances_
```

```
model.n_features_
```

```
for i in model.estimators_:
  i.fit(X_train, y_train)
  print('훈련 세트 정확도 : {:.3f}'.format(model.score(X_train, y_train)))
  print('테스트 세트 정확도 : {:.3f}'.format(model.score(X_train, y_train)))
```
