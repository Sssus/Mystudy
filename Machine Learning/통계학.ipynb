모분산을 알고 있을 경우에는 z분포를 사용하고
모분산을 모를 경우에는 t분포를 사용한다.

2종오류(베타) : 귀무가설이 틀렸는데 귀무가설을 채택할 경우

일반적으로 알파값을 많이쓴다. 95%의 신뢰수준으로 추정하려고 할 떄에 알파를 0.05로 한다.

'동일성검정'은 집단간의 모수의 특징이 동일함을 검정하는 것이다.
예를들어 집단1과 집단2의 평균이 동일한가? 이러한 문제를 검정하는 것으로 

모평균 동일성 검정에는 독립표본 t-test, 대응표본 t-test, 단일표본 t-test등으로 평균의 동일성을 검정하고 

모분산의 동일성 검정에는 2집단일 경우 F-test, 집단이 많을경우 Bartlett's Test가 쓰인다.

비율의 동일성 검정에는 proportion test 가 쓰이는데 요거는 통계량을 그냥 비율로 갖다쓰는걸로 기억한다.
