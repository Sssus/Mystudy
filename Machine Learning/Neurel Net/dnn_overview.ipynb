```
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
```

```
X,y=make_moons(n_samples=100,noise=0.25,random_state=3)
X_train,X_test,y_train,y_test = train_test_split(X,y,stratify=y,random_state=42)
```

```
mlp=MLPClassifier(solver='lbfgs',random_state=0).fit(X_train,y_train)
mlp
```

```
import numpy as np
import matplotlib.pyplot as plt
```

```
line=np.linspace(-3,3,100)
tanh_line=np.tanh(line)
relu_line=np.maximum(line,0)
sig_line=1/(1+np.exp(-line))
step_line=line.copy()
step_line[step_line<=0]=0
step_line[step_line>0]=1
```

```
matplotlib.rcParams['axes.unicode_minus']=False
```

```
plt.rcParams['figure.figsize']=(14,10)
```

```
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(2,2)
ax1.plot(line,tanh_line,label='tanh')
ax2.plot(line,np.maximum(line,0),label='relu',color='orange')
ax3.plot(line,sig_line,label='sigmoid',color='red')
ax4.plot(line,step_line,label='step',color='black')
```

```
plt.subplot(2,2,1)
plt.plot(sig_line)
plt.subplot(2,2,2)
plt.plot(line,tanh_line)
plt.subplot(2,2,3)
plt.plot(line,)
plt.subplot(2,2,4)
```
