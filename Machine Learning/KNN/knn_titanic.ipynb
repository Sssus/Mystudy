```
from sklearn.neighbors import KNeighborsClassifier
```

```
train.columns
```

```
sel=['PassengerId','Pclass','SibSp','Parch']

X_train=train[sel]
y_train=train['Survived']
X_test=test[sel]
```

```
model=KNeighborsClassifier()
model.fit(X_train,y_train)
```

```
pred=model.predict(X_test)
pred[:15]
```
